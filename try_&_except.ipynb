
*Smart Calculator*


```python
def add(a, b):
    return a + b

def sub(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    return a / b

def modulas(a, b):
    return a % b

def Square(a):
    return a ** 2

def percentage(a, b):
    return a/b * 100

def power(a, b):
    return a ** b

def cube(a):
    return a ** 3



while True:

    try:
        num1 = int(input("Enter the first number: "))
        num2 = int(input("Enter the second number: "))
    
        operations = input("Enter the operaions(+, -, *, /, %, Square, Percentage, Power, Cube): ")
    
    
        if operations == "+":
            print("The ans of num1 & num2 is: " , add(num1, num2))
    
        elif operations == "-":
           print("The ans of num1 & num2 is: " ,sub(num1, num2))
        
        
        elif operations == "*":
           print("The ans of num1 & num2 is: " , multiply(num1, num2))
    
        elif operations == "/":
           print("The ans of num1 & num2 is: " , divide(num1, num2))

        elif operations == "%":
           print("The ans of num1 & num2 is: " , modulas(num1, num2))

        elif operations == "Square":
           print("The ans of num1 is: ", Square(num1))
           print("The ans of num2 is: ", Square(num2))

        elif operations == "Percentage":
           print("The ans of num1 & num2 is: ", percentage(num1, num2))

        elif operations == "Power":
           print("The ans of num1 & num2 is: ", power(num1, num2))

        elif operations == "Cube":
           print("The ans of num1 is: ", cube(num1))
           print("The ans of num2 is: ", cube(num2))

        else:
           print("invalid operation/input")
    
    except ValueError:
        print("Enter the valid number only!!")
    
    except ZeroDivisionError: 
       print("Can't divide by zero!!")

    choice = input("Do you want to continue?? (yes/no): ")

    if choice == "no":
       print("Calculator closed")
       break
```
